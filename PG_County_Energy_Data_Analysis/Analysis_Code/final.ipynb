```python
df = df.dropna(subset=['Native Use','Cost'])
df=df[[ 'Place','Commodity ', 'Bill ID', 'Billing Period','Start Date', 'End Date', 'Days', 'Native Use', 'Cost','Common Use', 'Pay Amount']]
df.columns = [ 'Place','Commodity', 'Bill ID', 'Billing Period','Start Date', 'End Date', 'Days', 'Native Use', 'Cost','Common Use', 'Pay Amount']
df.head()
```

```python
df = df.drop(df[(df['Cost'] == 0) & (df['Native Use'] !=0)].index)
df = df.drop(df[(df['Cost'] != 0) & (df['Native Use'] ==0)].index)
df = df.drop(df[df['Cost'] < 0].index) 
df = df.drop(df[df['Native Use'] < 0].index)
```

```python
df.columns
```

```python
df = df.drop(df[df['Commodity'] == 'Water'].index)
df = df.drop(df[df['Commodity'] == 'Telecom'].index)
df = df.drop(df[df['Commodity'] == 'Water & Sewer'].index)
df['Commodity'].value_counts()
```

```python
df['Place'].value_counts()
```

```python
df=df.drop(df[df['Place'] =='Westphalia Community Center / Community Center'].index)
df.head()
```

```python
# Name standardization
df['Place'] = df['Place'].replace(['Palmer Park Community Center / Community Center'],'Palmer Park Community Center')
df['Place'] = df['Place'].replace(["Prince George's Publick Playhouse / Art Center"],"Prince George's Publick Playhouse")
df['Place'] = df['Place'].replace(["Laurel Beltsville Senior Activity Center/ Center"],"Laurel Beltsville Senior Activity Center")
df['Place'] = df['Place'].replace(["Southern Tech&Rec Complex/Athletic Center"],"Southern Tech&Rec Complex")
df['Place'] = df['Place'].replace(["Southern Tech & Rec Complex / Wellness Center"],"Southern Tech&Rec Complex")
df['Commodity'] = df['Commodity'].replace(["Water & Sewer"],"Water")
df.head()
```
